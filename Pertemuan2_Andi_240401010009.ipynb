# Latihan Pertemuan 2

```python
import pandas as pd
import numpy as np
# ① Muat Dataset Titanic
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)

# ② Eksplorasi Awal
print(df.shape, df.columns.tolist())
print(df.isnull().sum())

# ③ Filtering — Penumpang Wanita Kelas 1
wk1 = df[(df['Sex']=='female') &
(df['Pclass']==1)]
print(wk1['Survived'].mean(), wk1['Age'].mean())

# ④ Groupby — Tingkat Keselamatan per Kelas
hasil = df.groupby('Pclass').agg(
total=('PassengerId','count'),
selamat=('Survived','mean'),
).round(3)
print(hasil)
```

# Pertanyaan 1
## Berapa jumlah baris dan kolom dataset Titanic?

```python
# Pertanyaan 1
# Berapa jumlah baris dan kolom dataset Titanic?

df_shape = df.shape
baris = df_shape[0]
kolom = df_shape[1]
print('Jumlah baris:', baris)
print('Jumlah kolom:', kolom)
```

# Pertanyaan 2
## Kolom mana yang memiliki nilai kosong (missing)?


```python
# Pertanyaan 2
# Kolom mana yang memiliki nilai kosong (missing)?

missing_colm = df.isnull().sum()
missing_df = pd.DataFrame({
    'Jumlah Missing': missing_colm,
})

# Kolom yang memiliki missing
print(missing_df[missing_df['Jumlah Missing'] > 0])
```

# Pertanyaan 3
## Berapa persen penumpang yang selamat (Survived=1)?

```python
# Pertanyaan 3
# Berapa persen penumpang yang selamat (Survived=1)?

df_survived = df[df['Survived'] == 1]
survived_count = df_survived.shape[0]
total_count = df.shape[0]
print('Penumpang yang selamat sebanyak',(survived_count / total_count) * 100,'%')
```

# Pertanyaan 4
## Rata-rata usia penumpang wanita kelas 1 yang selamat?

```python
# Rata-rata usia penumpang wanita kelas 1 yang selamat?

df_female_class1_survived = df[(df['Sex'] == 'female') & (df['Pclass'] == 1) & (df['Survived'] == 1)]
df_female_class1_survived_age_series = df_female_class1_survived['Age']
print('Rata-rata berusia',df_female_class1_survived_age_series.mean(),'Tahun')
```

#Pertanyaan 5
## Kelas penumpang mana yang tingkat keselamatan tertinggi?

```python
# Kelas penumpang mana yang tingkat keselamatan tertinggi?

class_survive = df.groupby('Pclass')['Survived'].mean()
persen = class_survive.max()*100
print('Kelas dengan kelesamatan tertinggi adalah kelas ',class_survive.idxmax(),'dengan',persen,'% penumpang selamat')
```
