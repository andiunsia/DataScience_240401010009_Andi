# 5. Hapus baris duplikat

```python
df.drop_duplicates(inplace=True)
```

# 6. Normalisasi string

```python
df['kota'] = df['kota'].str.strip().str.title()
df['kondisi'] = df['kondisi'].str.strip().str.lower()
```

# 7. Imputasi missing values

```python
#median untuk numerik, modus untuk kategorik

df['luas_m2'] = df['luas_m2'].fillna(df['luas_m2'].median())
df['harga_juta'] = df['harga_juta'].fillna(df['harga_juta'].median())
df['kamar'] = df['kamar'].fillna(df['kamar'].mode()[0])
```

# 8. Tangani outlier

```python
#IQR Fence pada kolom harga_juta dan luas_m2

for col in ['harga_juta', 'luas_m2', 'tahun_bangun']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    df[col] = df[col].clip(Q1-1.5*IQR, Q3+1.5*IQR)
```

# 9. Validasi akhir

```python
assert df.isnull().sum().sum() == 0, 'Masih ada missing!'
assert df.duplicated().sum() == 0, 'Masih ada duplikat!'
print('Shape akhir:', df.shape)
```

# 10. Ekspor

```python
df.to_csv('housing_clean.csv', index=False)
print('Dataset bersih tersimpan!')
```

# **Kesimpulan**

**Data Cleaning dan Preprocessing**


Proses data cleaning pada dataset tujuannya adalah mengubah data mentah menjadi dataset yang bersih, konsisten, dan siap untuk analisis lebih lanjut. Mulai dari deteksi missing, penghapusan duplikat, normalisasi teks, hingga penanganan outlier, setiap langkah diterapkan dengan tepat sesuai karakteristik data. Proses ini menegaskan bahwa data cleaning adalah tahap krusial sebelum analisis atau pemodelan, karena kualitas data sangat memengaruhi validitas hasil akhir. Berikut tahapannya :

1.   Eksplorasi Awal (Data Inspection)

      Dari hasil inspeksi menggunakan info() dan describe()

2.   Penghapusan Duplikat

      Baris duplikat diidentifikasi dan dihapus menggunakan drop_duplicates(), memastikan setiap baris mewakili data unik.

3.   Normalisasi String (Konsistensi Format)

      Kolom kota diubah menjadi format Title Case (huruf pertama kapital) dan kolom kondisi menjadi lowercase menggunakan metode str.strip() dan str.title()/str.lower(). Hal ini menghilangkan spasi berlebih dan menyamakan format penulisan.

4.   Imputasi Missing Values

      Numerik: Kolom luas_m2 dan harga_juta diisi dengan nilai median (nilai tengah). Kolom kamar diisi dengan modus (nilai paling sering muncul).

      
5.   Penanganan Outlier

      Pada kolom harga_juta, luas_m2, dan tahun_bangun, nilai-nilai di luar batas lower dan upper (Q1 - 1.5×IQR dan Q3 + 1.5×IQR) dibatasi ke nilai batas tersebut agar dapat mempertahankan ukuran dataset (tidak menghapus baris) tetapi efektif terhadap pengaruh nilai yang ekstrem.

6.   Validasi Akhir

      Setelah seluruh tahapan, dilakukan validasi menggunakan assert untuk memastikan tidak ada lagi missing value dan baris duplikat.


7.   Ekspor Dataset Bersih

      Data bersih disimpan ke file housing_clean.csv untuk digunakan pada tahap analisis atau pemodelan selanjutnya.
